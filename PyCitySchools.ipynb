```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
total_schools = len(school_data_complete['school_name'].unique())
total_students = school_data_complete['Student ID'].count()
total_budget = school_data['budget'].sum()
average_math_score = school_data_complete['math_score'].mean()
average_reading_score = school_data_complete['reading_score'].mean()

student_passing_math = student_data.loc[student_data['math_score'] >= 70, :]['Student ID'].count()
math_passing_rate = (student_passing_math / total_students) * 100

passing_students_reading = student_data.loc[student_data['reading_score'] >= 70, :]['Student ID'].count()
passing_rate_reading = (passing_students_reading / total_students) * 100

overall_passing_rate = (math_passing_rate + passing_rate_reading) / 2
```

```python
display(total_schools,total_students,total_budget,average_math_score,average_reading_score,student_passing_math,math_passing_rate,passing_students_reading,passing_rate_reading,overall_passing_rate)
```

```python
district_summary = pd.DataFrame([[total_schools,total_students,total_budget,average_math_score,average_reading_score,student_passing_math,math_passing_rate,passing_students_reading,passing_rate_reading,overall_passing_rate]],
                               columns=['Total Schools','Total Students','Disctrict Budget','Avg Math Score','Avg Reading Score','Passing Math','Passing Math Rate','Passing Reading','Passing Reading Rate','Overall Passing Rate'])
```

```python
school_data_complete['passed_math'] = np.where(school_data_complete['math_score']>=70, 1, 0)
school_data_complete['passed_reading'] = np.where(school_data_complete['reading_score']>=70, 1, 0)
school_data_complete['passed_both'] = np.where((school_data_complete['math_score']>=70)&(school_data_complete['reading_score']>=70), 1, 0)

display(school_data_complete)
```

```python
school_summary=school_data_complete.groupby(['school_name','type']).agg(
    num_stud=('Student ID','count'),
    avg_math_score=('math_score','mean'),
    avg_reading_score=('reading_score','mean'),
    total_passed_math=('passed_math', 'sum'),
    total_passed_reading=('passed_reading', 'sum'),
    total_passed_both=('passed_both', 'sum')
).reset_index()

school_summary=school_summary.merge(school_data[['school_name','budget']],how='left',on=['school_name'])
school_summary['Budget per Student']=school_summary['budget']/school_summary['num_stud']
school_summary['% passed math']=school_summary['total_passed_math']/school_summary['num_stud']*100
school_summary['% passed reading']=school_summary['total_passed_reading']/school_summary['num_stud']*100
school_summary['% passed both']=school_summary['total_passed_both']/school_summary['num_stud']*100
                                                                             
school_summary = school_summary.drop(columns=['total_passed_math', 'total_passed_reading', 'total_passed_both'])                                                                             
```
