```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
display(school_data)
display(student_data)
display(school_data_complete)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data_complete['school_name'].unique())
total_students = school_data_complete['Student ID'].count()
total_budget = school_data['budget'].sum()
average_math_score = school_data_complete['math_score'].mean()
average_reading_score = school_data_complete['reading_score'].mean()

student_passing_math = student_data.loc[student_data['math_score'] >= 70, :]['Student ID'].count()
math_passing_rate = (student_passing_math / total_students) * 100

passing_students_reading = student_data.loc[student_data['reading_score'] >= 70, :]['Student ID'].count()
passing_rate_reading = (passing_students_reading / total_students) * 100

overall_passing_rate = (math_passing_rate + passing_rate_reading) / 2
```

```python
display(total_schools,total_students,total_budget,average_math_score,average_reading_score,student_passing_math,math_passing_rate,passing_students_reading,passing_rate_reading,overall_passing_rate)
```

```python
district_summary = pd.DataFrame([[total_schools,total_students,total_budget,average_math_score,average_reading_score,student_passing_math,math_passing_rate,passing_students_reading,passing_rate_reading,overall_passing_rate]],
                               columns=['Total Schools','Total Students','Disctrict Budget','Avg Math Score','Avg Reading Score','Passing Math','Passing Math Rate','Passing Reading','Passing Reading Rate','Overall Passing Rate'])
```

```python
display(district_summary)
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
display(school_data_complete.head(2))
```

```python
school_data_complete['passed_math'] = np.where(school_data_complete['math_score']>=70, 1, 0)
school_data_complete['passed_reading'] = np.where(school_data_complete['reading_score']>=70, 1, 0)
school_data_complete['passed_both'] = np.where((school_data_complete['math_score']>=70)&(school_data_complete['reading_score']>=70), 1, 0)

display(school_data_complete)
```

```python
school_summary=school_data_complete.groupby(['school_name','type']).agg(
    num_stud=('Student ID','count'),
    avg_math_score=('math_score','mean'),
    avg_reading_score=('reading_score','mean'),
    total_passed_math=('passed_math', 'sum'),
    total_passed_reading=('passed_reading', 'sum'),
    total_passed_both=('passed_both', 'sum')
).reset_index()

school_summary=school_summary.merge(school_data[['school_name','budget']],how='left',on=['school_name'])
school_summary['Budget per Student']=school_summary['budget']/school_summary['num_stud']
school_summary['% passed math']=school_summary['total_passed_math']/school_summary['num_stud']*100
school_summary['% passed reading']=school_summary['total_passed_reading']/school_summary['num_stud']*100
school_summary['% passed both']=school_summary['total_passed_both']/school_summary['num_stud']*100
                                                                             
school_summary = school_summary.drop(columns=['total_passed_math', 'total_passed_reading', 'total_passed_both'])                                                                             
```

```python
display(school_summary)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
students_9th  = school_data_complete.loc[school_data_complete['grade'] == '9th',]
students_10th = school_data_complete.loc[school_data_complete['grade'] == '10th',]
students_11th = school_data_complete.loc[school_data_complete['grade'] == '11th',]
students_12th = school_data_complete.loc[school_data_complete['grade'] == '12th',]

students_9th_school_count  = students_9th.groupby('school_name')['grade'].count()
students_10th_school_count = students_10th.groupby('school_name')['grade'].count()
students_11th_school_count = students_11th.groupby('school_name')['grade'].count()
students_12th_school_count = students_12th.groupby('school_name')['grade'].count()

math_9th_school  = students_9th.groupby('school_name')['math_score'].sum() / students_9th_school_count
math_10th_school = students_10th.groupby('school_name')['math_score'].sum() / students_10th_school_count
math_11th_school = students_11th.groupby('school_name')['math_score'].sum() / students_11th_school_count
math_12th_school = students_12th.groupby('school_name')['math_score'].sum() / students_12th_school_count

math_scores_grade_school = pd.concat([math_9th_school, 
                                    math_10th_school,
                                    math_11th_school,
                                    math_12th_school],
                                    axis = 1)

math_scores_grade_school = math_scores_grade_school.rename(columns={0 :'9th', 1 : '10th', 2 : '11th', 3 : '12th'})


math_scores_grade_school
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
students_9th  = school_data_complete.loc[school_data_complete['grade'] == '9th',]
students_10th = school_data_complete.loc[school_data_complete['grade'] == '10th',]
students_11th = school_data_complete.loc[school_data_complete['grade'] == '11th',]
students_12th = school_data_complete.loc[school_data_complete['grade'] == '12th',]

students_9th_school_count  = students_9th.groupby('school_name')['grade'].count()
students_10th_school_count = students_10th.groupby('school_name')['grade'].count()
students_11th_school_count = students_11th.groupby('school_name')['grade'].count()
students_12th_school_count = students_12th.groupby('school_name')['grade'].count()

reading_9th_school  = students_9th.groupby('school_name')['reading_score'].sum() / students_9th_school_count
reading_10th_school = students_10th.groupby('school_name')['reading_score'].sum() / students_10th_school_count
reading_11th_school = students_11th.groupby('school_name')['reading_score'].sum() / students_11th_school_count
reading_12th_school = students_12th.groupby('school_name')['reading_score'].sum() / students_12th_school_count

reading_scores_grade_school = pd.concat([reading_9th_school, 
                                    reading_10th_school,
                                    reading_11th_school,
                                    reading_12th_school],
                                    axis = 1)

reading_scores_grade_school = reading_scores_grade_school.rename(columns={0 :'9th', 1 : '10th', 2 : '11th', 3 : '12th'})


reading_scores_grade_school
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
school_data_complete['budget_per_student']=school_data_complete['budget']/school_data_complete['size']
groups=['>$577','>$625 to $638','>$638 to $650','>$650']

school_data_complete['budget_group']=pd.qcut(school_data_complete['budget_per_student'], q=4, labels=groups)

size_summary=school_data_complete.groupby(['budget_group']).agg(
    num_stud=('Student ID','count'),
    avg_math_score=('math_score','mean'),
    avg_reading_score=('reading_score','mean'),
    total_passed_math=('passed_math', 'sum'),
    total_passed_reading=('passed_reading', 'sum'),
    total_passed_both=('passed_both', 'sum')
).reset_index()

#school_summary=school_summary.merge(school_data[['type']],how='left',on=['type'])
size_summary['Average Math Score']=size_summary['avg_math_score']
size_summary['Average Reading Score']=size_summary['avg_reading_score']
size_summary['% passed math']=size_summary['total_passed_math']/size_summary['num_stud']*100
size_summary['% passed reading']=size_summary['total_passed_reading']/size_summary['num_stud']*100
size_summary['% passed both']=size_summary['total_passed_both']/size_summary['num_stud']*100
                                                                             
size_summary = size_summary.drop(columns=['num_stud','total_passed_math', 'total_passed_reading', 'total_passed_both'])   

display(size_summary)

```
